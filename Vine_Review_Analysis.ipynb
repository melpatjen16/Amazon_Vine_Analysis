```python
vine_data_df = pd.read_csv('Vine_Table.csv')
vine_data_df.head()
```

```python
filter_vine_data = vine_data_df[vine_data_df['total_votes'] >= 20]
filter_vine_data.head()
```

```python
vine_df = filter_vine_data[filter_vine_data.helpful_votes/ filter_vine_data.total_votes >= 0.5]
vine_df.head()
```

```python
vine_reviews = vine_df[vine_df['vine'] == 'Y']
vine_reviews.head()
```

```python
unpaid_vine_reviews = vine_df[vine_df['vine'] == 'N']
unpaid_vine_reviews.head()
```
